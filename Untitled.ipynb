```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def render_stereo_img(imgL, imgR, gray=True):
    """Handy function to plot images"""
    if gray:
        plt.subplot(121),plt.imshow(cv2.cvtColor(imgL, cv2.COLOR_BGR2GRAY), cmap = 'gray')
        plt.title('Left'), plt.xticks([]), plt.yticks([])
        plt.subplot(122),plt.imshow(cv2.cvtColor(imgR, cv2.COLOR_BGR2GRAY), cmap = 'gray')
        plt.title('Right'), plt.xticks([]), plt.yticks([])
    else:
        plt.subplot(121),plt.imshow(cv2.cvtColor(imgL, cv2.COLOR_BGR2RGB))
        plt.title('Left'), plt.xticks([]), plt.yticks([])
        plt.subplot(122),plt.imshow(cv2.cvtColor(imgR, cv2.COLOR_BGR2RGB))
        plt.title('Right'), plt.xticks([]), plt.yticks([])
```

```python
imgL = cv2.imread('./example_data/example_data/Drill_Left_10_n1p190_d0p784.png')
imgR = cv2.imread('./example_data/example_data/Drill_Right_10_n1p190_d0p784.png')

render_stereo_img(imgL, imgR, gray=True)
```

```python
def stereo_compute(imgL, imgR, numDisparities=16, blockSize=15):
    stereo = cv2.StereoBM_create(numDisparities=numDisparities, blockSize=blockSize)
    disparity = stereo.compute(imgL, imgR)
    return stereo, disparity
```

```python
imgL_gray = cv2.cvtColor(imgL.copy(), cv2.COLOR_BGR2GRAY)
imgR_gray = cv2.cvtColor(imgR.copy(), cv2.COLOR_BGR2GRAY)
s, d = stereo_compute(imgL_gray, imgR_gray, numDisparities=16, blockSize=21)
plt.imshow(d, cmap='gray')
```
